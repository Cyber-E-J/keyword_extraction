```python

def open_file(filename):
    file = open(filename,"r")
    lines = file.readlines()
    dialogues = []
    output = open("gold_val_de.txt","w")
    for line in lines:
        dialogue_in = line.find("summary_de")
        if (dialogue_in!=-1):
            dialogues.append(line.strip()[15:-2])
            output.write(line.strip()[15:-2])
            output.write('\n')
    return dialogues

open_file("val.json")
```
